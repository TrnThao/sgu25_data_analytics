## Bài toán 1: Thực hiện các nhiệm vụ trong bài toán để làm quen với các công cụ trực quan hóa dữ liệu.

Dữ liệu thực hiện là dữ liệu về giá nhà lấy từ https://www.kaggle.com/datasets/thomasnibb/amsterdam-
house-price-prediction

## Nhiệm vụ 1:
1. Chuẩn bị dữ liệu cho trực quan hóa dữ liệu

```python
import pandas as pd
houseprices_data = pd.read_csv("data/HousingPrices-Amsterdam-August-2021.csv")
houseprices_data = houseprices_data[['Zip', 'Price', 'Area', 'Room']]
# Create a PriceperSqm variable based on the Price and Area variables:
houseprices_data['PriceperSqm'] = houseprices_data['Price']
houseprices_data['Area']
```

## 2. Trực quan hóa dữ liệu với thư viện Matplotlib

```python
import matplotlib.pyplot as plt

houseprices_sorted = houseprices_data.sort_values('Price', ascending =False)
houseprices_sorted.head()
# case 1: basic
plt.figure(figsize= (12,6))
x = houseprices_sorted['Zip'][0:10]
y = houseprices_sorted['Price'][0:10]
plt.bar(x,y)
plt.show()
# case 2: advanced 1
plt.figure(figsize= (12,6))
plt.bar(x,y)
plt.title('Top 10 Areas with the highest house prices', fontsize=15)
plt.xlabel('Zip code', fontsize = 12)
plt.xticks(fontsize=10)
plt.ylabel('House prices in millions', fontsize=12)
plt.yticks(fontsize=10)
plt.show()
# case 3: advanced 2
fig, ax = plt.subplots(figsize=(40,18))
x = houseprices_sorted['Zip'][0:10]
y = houseprices_sorted['Price'][0:10]
y1 = houseprices_sorted['PriceperSqm'][0:10]
plt.subplot(1,2,1)
plt.bar(x,y)
plt.xticks(fontsize=17)
plt.ylabel('House prices in millions', fontsize=25)
plt.yticks(fontsize=20)
plt.title('Top 10 Areas with the highest house prices',
fontsize=25)
plt.subplot(1,2,2)
plt.bar(x,y1)
plt.xticks(fontsize=17)
plt.ylabel('House prices per sqm', fontsize=25)
plt.yticks(fontsize=20)
plt.title('Top 10 Areas with the highest house prices per sqm',
fontsize=25)
plt.show()
```

## 3. Trực quan hóa dữ liệu với thư viện Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
# case 1: basic
plt.figure(figsize= (12,6))
data = houseprices_sorted[0:10]
sns.barplot(data= data, x= 'Zip',y = 'Price')
# case 2: advanced 1
plt.figure(figsize= (12,6))
data = houseprices_sorted[0:10]
ax = sns.barplot(data= data, x= 'Zip',y = 'Price')
ax.set_xlabel('Zip code',fontsize = 15)
ax.set_ylabel('House prices in millions', fontsize = 15)
ax.set_title('Top 10 Areas with the highest house prices', fontsize= 20)
# case 3: view multiple perspectives at once
fig, ax = plt.subplots(1, 2,figsize=(40,18))
data = houseprices_sorted[0:10]
sns.set(font_scale = 3)
ax1 = sns.barplot(data= data, x= 'Zip',y = 'Price', ax = ax[0])
ax1.set_xlabel('Zip code')
ax1.set_ylabel('House prices in millions')
ax1.set_title('Top 10 Areas with the highest house prices')
ax2 = sns.barplot(data= data, x= 'Zip',y = 'PriceperSqm',
ax=ax[1])
ax2.set_xlabel('Zip code')
ax2.set_ylabel('House prices per sqm')
ax2.set_title('Top 10 Areas with the highest price per sqm')
```

## Bài tập thực hành 1

## 1. Tải và đọc dữ liệu

```python
import pandas as pd

# Đọc dữ liệu
df = pd.read_csv("winequality-red.csv")
df.head()
```

## 2. Kiểm tra thông tin và mô tả dữ liệu

```python
df.info()
df.describe()
df['quality'].value_counts()
```

## 3. Trực quan hóa dữ liệu cơ bản
Phân bố chất lượng rượu

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.figure(figsize=(8,5))
sns.countplot(x='quality', data=df, palette='Set2')
plt.title("Phân bố chất lượng rượu đỏ")
plt.xlabel("Chất lượng (Quality)")
plt.ylabel("Số lượng mẫu")
plt.show()
```

## 4. Phân bố của các đặc trưng (features)

```python
df.hist(bins=15, figsize=(15,10), color='salmon', edgecolor='black')
plt.suptitle("Phân bố các thuộc tính hoá học của rượu đỏ", fontsize=16)
plt.show()
```

## 5. Ma trận tương quan giữa các thuộc tính

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Ma trận tương quan giữa các thuộc tính")
plt.show()
```

## 6. So sánh từng yếu tố với chất lượng

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='quality', y='alcohol', data=df, palette='Set3')
plt.title("Mối quan hệ giữa độ cồn và chất lượng rượu")
plt.show()

plt.figure(figsize=(12,6))
sns.boxplot(x='quality', y='volatile acidity', data=df, palette='coolwarm')
plt.title("Mối quan hệ giữa độ chua dễ bay hơi và chất lượng rượu")
plt.show()
```

## 7. Pairplot – xem toàn cảnh tương quan

```python
sns.pairplot(df[['alcohol', 'volatile acidity', 'citric acid', 'density', 'quality']],
             hue='quality', palette='husl')
plt.suptitle("Biểu đồ pairplot giữa các thuộc tính và chất lượng", y=1.02)
plt.show()
```

## 8. Gợi ý nâng cao (tuỳ chọn)

```python
df['quality_label'] = df['quality'].apply(lambda q: 'Tốt' if q >= 7 else ('Kém' if q <= 4 else 'Trung bình'))
sns.countplot(x='quality_label', data=df, palette='Set1')
plt.title("Phân loại chất lượng rượu: Tốt / Trung bình / Kém")
plt.show()
```

## 1. Chuẩn bị môi trường

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

df = pd.read_csv("diabetes.csv")
df.head()

```

## 2. Khám phá sơ bộ dữ liệu

```python
df.info()
df.describe()
df['Outcome'].value_counts()
```

## 3. Phân bố dữ liệu đầu ra (Outcome)

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Outcome', data=df, palette='Set2')
plt.title("Phân bố người mắc và không mắc tiểu đường")
plt.xlabel("Kết quả chẩn đoán (0 = Không, 1 = Có)")
plt.ylabel("Số lượng mẫu")
plt.show()
```

## 4. Thống kê mô tả cho từng biến theo Outcom

```python
plt.figure(figsize=(10,6))
df.groupby('Outcome').mean().T.plot(kind='bar', figsize=(12,6))
plt.title("Giá trị trung bình các đặc trưng theo nhóm Outcome")
plt.ylabel("Giá trị trung bình")
plt.show()
```

## 5. Biểu đồ phân bố (Histogram / KDE)

```python
plt.figure(figsize=(8,5))
sns.histplot(df[df['Outcome']==1]['Glucose'], color='red', label='Mắc bệnh', kde=True)
sns.histplot(df[df['Outcome']==0]['Glucose'], color='green', label='Không mắc', kde=True)
plt.title("Phân bố mức Glucose theo Outcome")
plt.legend()
plt.show()
```

## 6. Ma trận tương quan giữa các đặc trưng

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Ma trận tương quan giữa các thuộc tính trong dữ liệu tiểu đường")
plt.show()
```

## 7. Mối quan hệ từng biến với Outcome (Boxplot / Violinplot)

```python
features = ['Glucose', 'BMI', 'Age', 'Insulin', 'BloodPressure']

for f in features:
    plt.figure(figsize=(8,5))
    sns.boxplot(x='Outcome', y=f, data=df, palette='Set3')
    plt.title(f'Mối quan hệ giữa {f} và Outcome')
    plt.show()
```

## 8. Pairplot – xem tương quan tổng quát

```python
sns.pairplot(df[['Glucose', 'BMI', 'Age', 'BloodPressure', 'Outcome']],
             hue='Outcome', palette='husl')
plt.suptitle("Biểu đồ Pairplot giữa các biến và tình trạng tiểu đường", y=1.02)
plt.show()
```

## 9. Tạo phân loại cơ bản cho tuổi

```python
df['AgeGroup'] = pd.cut(df['Age'], bins=[20,30,40,50,60,70,80], labels=['20s','30s','40s','50s','60s','70s'])
sns.countplot(data=df, x='AgeGroup', hue='Outcome', palette='Set1')
plt.title("Phân bố bệnh tiểu đường theo nhóm tuổi")
plt.show()
```

## + Thực hiện EDA trên tập dữ liệu mua sắm tại siêu thị. Tập dữ liệu lấy từ
https://www.kaggle.com/code/rajatkumar30/eda-online-retail

## 1. Chuẩn bị môi trường & đọc dữ liệu

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Đọc dữ liệu
df = pd.read_excel("Online Retail.xlsx")
df.head()
```

## 2. Khám phá sơ bộ dữ liệu

```python
df.info()
df.describe()
df.shape
df.isnull().sum()
```

## 3. Làm sạch dữ liệu cơ bản

```python
# Loại bỏ hàng bị thiếu CustomerID
df = df.dropna(subset=['CustomerID'])

# Loại bỏ giá hoặc số lượng âm
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]

# Thêm cột tổng tiền
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

## 4. Thống kê tổng quan

```python
print("Số lượng khách hàng:", df['CustomerID'].nunique())
print("Số lượng sản phẩm:", df['StockCode'].nunique())
print("Số lượng quốc gia:", df['Country'].nunique())
```

## 5. Phân bố giao dịch theo quốc gia

```python
plt.figure(figsize=(12,6))
country_data = df['Country'].value_counts().head(10)
sns.barplot(x=country_data.index, y=country_data.values, palette="coolwarm")
plt.title("Top 10 quốc gia có số lượng giao dịch cao nhất")
plt.ylabel("Số lượng giao dịch")
plt.xlabel("Quốc gia")
plt.xticks(rotation=45)
plt.show()
```

## 6. Top sản phẩm bán chạy nhất

```python
plt.figure(figsize=(12,6))
top_products = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=top_products.values, y=top_products.index, palette="mako")
plt.title("Top 10 sản phẩm bán chạy nhất")
plt.xlabel("Tổng số lượng bán")
plt.ylabel("Sản phẩm")
plt.show()
```

## 7. Phân tích theo thời gian

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Month'] = df['InvoiceDate'].dt.month

plt.figure(figsize=(10,5))
sns.countplot(x='Month', data=df, palette='crest')
plt.title("Số lượng giao dịch theo tháng")
plt.xlabel("Tháng")
plt.ylabel("Số lượng giao dịch")
plt.show()
```

## 8. Doanh thu theo quốc gia

```python
revenue = df.groupby('Country')['TotalPrice'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=revenue.values, y=revenue.index, palette="viridis")
plt.title("Top 10 quốc gia có doanh thu cao nhất")
plt.xlabel("Doanh thu (£)")
plt.ylabel("Quốc gia")
plt.show()
```

## 9. Doanh thu theo sản phẩm

```python
top_rev = df.groupby('Description')['TotalPrice'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_rev.values, y=top_rev.index, palette="rocket")
plt.title("Top 10 sản phẩm mang lại doanh thu cao nhất")
plt.xlabel("Doanh thu (£)")
plt.ylabel("Sản phẩm")
plt.show()
```

## 10. Phân tích hành vi khách hàng — RFM

```python
snapshot_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'InvoiceNo': 'count',
    'TotalPrice': 'sum'
})
rfm.rename(columns={'InvoiceDate':'Recency','InvoiceNo':'Frequency','TotalPrice':'Monetary'}, inplace=True)
rfm.head()
rfm.describe()
sns.pairplot(rfm)
plt.suptitle("Phân tích RFM khách hàng", y=1.02)
plt.show()
```
